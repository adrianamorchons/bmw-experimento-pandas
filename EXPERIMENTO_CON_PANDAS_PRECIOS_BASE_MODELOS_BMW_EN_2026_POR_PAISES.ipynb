15. DETECCIÓN DE OUTLIERS CON IQR

```python
#Detectar posibles outliers
import matplotlib.pyplot as plt
import numpy as np


for column in df.select_dtypes(include=np.number).columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)][['Modelo', column]]
    if not outliers.empty:
        print(f"\nPosibles outliers en la columna '{column}':")
        print(outliers)
    else:
        print(f"\nNo se detectaron outliers en la columna '{column}'.")
```

15.1 Representación de outliers con gráfico de barras: datos que se pueden considerar atípicos

```python
#Representación de los outliers con gráficos de barras
#Solamente se muestran los registros más altos, los que superan los límites calculados
import matplotlib.pyplot as plt
import numpy as np

#Iteración sobre las columnas numéricas
for column in df.select_dtypes(include=np.number).columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound= Q1-1.5*IQR
    upper_bound= Q3+1.5*IQR

    #Filtrar solamente las filas de outliers que superan el límite superior
    outliers_superior = df[df[column] > upper_bound]

    if not outliers_superior.empty:
        print(f"\nGráfico de Outliers (solo superiores) en la columna '{column}':")
        plt.figure(figsize=(8, 5))
        plt.bar(outliers_superior['Modelo'], outliers_superior[column], color='skyblue')
        plt.xlabel('Modelo')
        plt.ylabel(f'Valor en {column}')
        plt.title(f'Outliers (superiores) detectados en {column}')
        plt.xticks(rotation=360, ha='right')
        plt.tight_layout()
        plt.show()
    else:
        print(f"\nNo se detectaron outliers superiores en la columna '{column}'.")

```

16.1. Representación gráfica del outlier del XM

```python
import matplotlib.pyplot as plt
import seaborn as sns

paises_grafico = []
distancias_grafico = []

# Asegurarse de que los nombres de las columnas estén limpios
df.columns = df.columns.str.strip()

paises_a_analizar = ['España','Alemania','Francia','Portugal','Italia']

#Calcular métricas iterando cada país
for pais in paises_a_analizar:
    # Obtener los precios de la columna del país actual
    country_prices = df[pais]

    Q1 = country_prices.quantile(0.25)
    Q3 = country_prices.quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR

    # Filtrar el modelo 'XM' del DataFrame completo
    df_xm_row = df[df['Modelo'].str.contains('XM', case=False, na=False)]

    if not df_xm_row.empty:
        precio_xm = df_xm_row[pais].values[0] # Obtener el precio del XM para este país

        # Verificar si el precio del XM es un outlier superior
        if precio_xm > upper_bound:
            distancia_veces_iqr = (precio_xm - Q3) / IQR

            # Guardar datos del outlier
            paises_grafico.append(pais)
            distancias_grafico.append(distancia_veces_iqr)

#Creación de gráfico (barras horizontales para facilitar la lectura)
plt.figure(figsize=(8, 5))

# Solo crear el gráfico si hay datos de outliers para mostrar
if paises_grafico:
    # Añadir hue=paises_grafico and legend=False
    sns.barplot(x=distancias_grafico, y=paises_grafico, palette='viridis', hue=paises_grafico, legend=False)

    #Etiquetas de datos para ver valor exacto al final de cada barra:
    for i, valor in enumerate(distancias_grafico):
        plt.text(valor + 0.1, i, f"{valor:.2f}x", va='center', fontweight='bold')

    #Personalización
    plt.axvline(x=1.5, color='red', linestyle='--', linewidth=1.5, label='Umbral Mínimo de Outlier (1.5x IQR)')
    plt.title('¿Cuántas veces supera el BMW XM el IQR desde el Q3?', fontsize=10, pad=15)
    plt.xlabel('Distancia en veces el IQR (desde Q3)', fontsize=10)
    plt.ylabel('País', fontsize=11)
    plt.legend(loc='lower right')
    plt.grid(axis='x', linestyle=':', alpha=0.6)

    plt.tight_layout()
    plt.show()
else:
    print("No se encontraron outliers del modelo XM por encima del umbral en los países seleccionados para graficar.")
```
