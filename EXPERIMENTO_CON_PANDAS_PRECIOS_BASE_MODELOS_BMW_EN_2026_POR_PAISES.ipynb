15. DETECCIÓN DE OUTLIERS CON IQR

```python
#Detectar posibles outliers
import matplotlib.pyplot as plt
import numpy as np


for column in df.select_dtypes(include=np.number).columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)][['Modelo', column]]
    if not outliers.empty:
        print(f"\nPosibles outliers en la columna '{column}':")
        print(outliers)
    else:
        print(f"\nNo se detectaron outliers en la columna '{column}'.")
```

15.1 Representación de outliers con gráfico de barras: datos que se pueden considerar atípicos

16. ANÁLISIS DEL MODELO XM


***NOTA**: se usa el método .isin() para filtrar datos mediante lista de valores, y solo selecciona las filas que coinciden con los elementos que indiquemos.*

```python
#Objetivo: analizar el impacto de este modelo 'minoritario' como outlier

import pandas as pd

#Filtramos los cinco países del dataset y registros del XM en cada uno
#Objetivo: XM

paises_objetivo = ['España','Alemania','Francia','Portugal','Italia']

# Quitar espacios en blanco que pueda haber
df.columns = df.columns.str.strip()

print("Análisis de desviación del BMW XM por país:")
print("=" * 60)

# Recorrer lista de países
for pais in paises_objetivo:

    # Datos de precios incluyendo la columna "Modelo"
    df_country_prices = df[['Modelo', pais]].copy()

    # IQR sobre los precios actuales por país
    Q1 = df_country_prices[pais].quantile(0.25)
    Q3 = df_country_prices[pais].quantile(0.75)
    IQR = Q3 - Q1

    # Outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Encontrar XM entre los datos
    df_xm = df_country_prices[df_country_prices['Modelo'].str.contains('XM', case=False, na=False)]

    if not df_xm.empty:
        precio_xm = df_xm[pais].values[0] # Get the price of the 'XM' model for this country

        # Comprobar si el precio del modelo XM es un outlier
        if precio_xm > upper_bound:
            distancia_veces_iqr = (precio_xm - Q3) / IQR

            print(f"País: {pais}")
            print(f"  -> Q3 del mercado: {Q3:.2f} € | Umbral Outlier: {upper_bound:.2f} €")
            print(f"  -> Precio del BMW XM: {precio_xm:.2f} €")
            print(f"  -> Distancia estadística: ¡Supera el Q3 en {distancia_veces_iqr:.2f} veces el IQR!\n")
        else:
            print(f"País: {pais} -> El BMW XM no es un outlier superior en este país. Precio: {precio_xm:.2f} €\n")
    else:
        print(f"País: {pais} -> No se encontraron registros del BMW XM en este país.\n")
```

16.1. Representación gráfica del outlier del XM

```python
import matplotlib.pyplot as plt
import seaborn as sns

paises_grafico = []
distancias_grafico = []

# Asegurarse de que los nombres de las columnas estén limpios
df.columns = df.columns.str.strip()

paises_a_analizar = ['España','Alemania','Francia','Portugal','Italia']

#Calcular métricas iterando cada país
for pais in paises_a_analizar:
    # Obtener los precios de la columna del país actual
    country_prices = df[pais]

    Q1 = country_prices.quantile(0.25)
    Q3 = country_prices.quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR

    # Filtrar el modelo 'XM' del DataFrame completo
    df_xm_row = df[df['Modelo'].str.contains('XM', case=False, na=False)]

    if not df_xm_row.empty:
        precio_xm = df_xm_row[pais].values[0] # Obtener el precio del XM para este país

        # Verificar si el precio del XM es un outlier superior
        if precio_xm > upper_bound:
            distancia_veces_iqr = (precio_xm - Q3) / IQR

            # Guardar datos del outlier
            paises_grafico.append(pais)
            distancias_grafico.append(distancia_veces_iqr)

#Creación de gráfico (barras horizontales para facilitar la lectura)
plt.figure(figsize=(8, 5))

# Solo crear el gráfico si hay datos de outliers para mostrar
if paises_grafico:
    # Añadir hue=paises_grafico and legend=False
    sns.barplot(x=distancias_grafico, y=paises_grafico, palette='mako', hue=paises_grafico, legend=False)

    #Etiquetas de datos para ver valor exacto al final de cada barra:
    for i, valor in enumerate(distancias_grafico):
        plt.text(valor + 0.1, i, f"{valor:.2f}x", va='center', fontweight='bold')

    #Personalización
    plt.axvline(x=1.5, color='red', linestyle='--', linewidth=1.5, label='Umbral Mínimo de Outlier (1.5x IQR)')
    plt.title('¿Cuántas veces supera el BMW XM el IQR desde el Q3?', fontsize=10, pad=15)
    plt.xlabel('Distancia en veces el IQR (desde Q3)', fontsize=10)
    plt.ylabel('País', fontsize=11)
    plt.legend(loc='lower right')
    plt.grid(axis='x', linestyle=':', alpha=0.6)

    plt.tight_layout()
    plt.show()
else:
    print("No se encontraron outliers del modelo XM por encima del umbral en los países seleccionados para graficar.")
```
